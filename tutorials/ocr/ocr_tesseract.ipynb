```python
import pytesseract
from PIL import Image
from googletrans import Translator
```

### OCR using Tesseract 

#### English Text Example

```python
en_image = Image.open('english_image_w_text.jpg')

text = pytesseract.image_to_string(en_image, lang='eng', config='--psm 6')
print(text)

new_size = (200, 150)
en_shrunken_image = en_image.resize(new_size)
en_shrunken_image
```

#### Chinese Text Example

```python
chi_image = Image.open('chinese_image_w_text.jpg')
chi_text = pytesseract.image_to_string(chi_image, lang='chi_sim' , config='--psm 6')
print(chi_text)

new_size = (200, 150)
chi_shrunken_image = chi_image.resize(new_size)
chi_shrunken_image
```

#### Translation to English

```python
# Create a Translator object
translator = Translator()
# Detect the source language (optional)
chi_text = chi_text.replace('\n', ' ')
detected_language = translator.detect(chi_text).lang
# Translate to English
translation = translator.translate(chi_text, src=detected_language, dest='en')

# Print the translation
print(f"Original Text ({detected_language}): {chi_text}")
print(f"Translation (English): {translation.text}")
```
