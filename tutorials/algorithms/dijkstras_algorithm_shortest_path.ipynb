```python
import networkx as nx
import matplotlib.pyplot as plt
```

##  Shortest Path Estimation using Dijkstra's Algorithm

Dijkstra's Algorithm is a graph algorithm that finds the shortest path from a given initial node to all other nodes in a weighted graph. 

To find shortest path, the algorithm initializes distances from the initial node (A) to all other nodes as infinity, and the distance to the A as 0. It then selects the unvisited node with the smallest distance (A) and marks it as visited. From A, it updates the distance of adjacent unvisited nodes (e.g., B, C, D, E) by considering the combined distance through A and selecting the minimum. These steps are repeated until all nodes are visited. 

Considerations when using Dijkstra's Algorithm: 
- Dijkstra's Algorithm uses priority queue to order adjacent unvisited nodes based on their distance to the source node (smallest first). 
- Dijkstra's Algorithm performs a relaxation step when updating the distances of adjacent unvisited nodes. It considers the combined weight through the current node and compares it with the existing distance to the adjacent node. If the combined weight is smaller, the distance to the adjacent node is updated to reflect the smaller distance value.

- Dijkstra's Algorithm is considered a greedy algorithm; it always chooses the node with the smallest distance at each step. This ensures that distance calculations for visited notes are held constant in subsequent iterations. 

- Dijkstra's Algorithm does not handle negative distances/weights.


What do the weights represent? 
- Weights represent the "cost" or "distance" associated with each edge in the graph. In transportation networks weights could represent physical distance, travel time, or monetary cost required to traverse from one geolocation node to another. In social networks, weights could represent number of interactions, the strength of a relationship, or level of influence between individual and/or social group nodes. In communication networks, weights could represent latency, bandwidth, or cost of transmission between device nodes.

```python
def create_weighted_digraph(graph_matrix):
    # Create a directed graph
    G = nx.DiGraph()
    # Adding edges with weights from the graph_matrix
    for edge_data in graph_matrix:
        source, target, weight = edge_data
        G.add_edge(source, target, weight=weight)
    return G


def find_shortest_path(graph, source, target):
    # Calculate the shortest path from source to target using Dijkstra's algorithm
    shortest_path = nx.shortest_path(graph, source=source, target=target, weight='weight')
    return shortest_path


def plot_weighted_digraph(G):
    # Draw the graph
    pos = nx.spring_layout(G)
    nx.draw(G, pos, with_labels=True, node_size=1500, node_color='skyblue', font_size=10, font_weight='bold')
    edge_labels = nx.get_edge_attributes(G, 'weight')
    nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels)
    # Show the plot
    plt.show()
```

```python
# Define the graph using a matrix representation
graph_matrix = [
    ('A', 'B', 34),
    ('A', 'C', 112),
    ('B', 'C', 15),
    ('B', 'D', 10),
    ('C', 'D', 38),
    ('D', 'E', 71),
    ('C', 'E', 9)
]
```

```python
# Create the graph
G = create_weighted_digraph(graph_matrix)

# Find the shortest path from node 'A' to node 'E'
shortest_path = find_shortest_path(G, 'A', 'E')
print("Shortest path:", shortest_path)
```

```python
# Generate and show the plot for the graph visualization
plot_weighted_digraph(G)
```
